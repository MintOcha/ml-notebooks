## Filtering data

Here's one of the most powerful features of DataFrames--being able to quickly work with large chunks of data. If you had to do this with for loops, it'd be a bit of pain to filter everything out item by item, not to mention having to reconstruct your lists one by one.

```python
attack = data.loc["Bulbasaur": "Mew", "attack" ]
attack
```

```python
attack > 100
```

```python
condition = attack > 100
attack[condition]
```

```python
attack[attack > 100]
```

## More Filtering

Take a look at what's being done, and try to figure it out, particularly when it comes to the two-condition criteria!

```python
# Can you figure out what's being done in the below code?

pokeAll = pd.read_csv("https://raw.githubusercontent.com/iamamangosteen/pythonforbusiness/master/pokedex_(Update_2021.04).csv", index_col = "name")
#import the data, set the index_col

criteria = (pokeAll["generation"] == 1) & (pokeAll["total_points"] > 100)
#define two filtering conditions (generation==1 and total_points>100) and save them in the criteria variable
#this is the first time we see column-shorthand notation

pokeAll["all_defense"] = pokeAll["defense"] + pokeAll["sp_defense"]
#create a new column called all_defense that is created by adding the defense and special defense columns

pokeAll[criteria][["defense","sp_defense", "all_defense"]]
#apply the filter criteria, then use column-shorthand notation to display only 3 columns
```
